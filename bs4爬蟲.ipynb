```
from urllib.request import urlopen
from bs4 import BeautifulSoup

page=1

while True:
  url = "https://tabelog.com/rstLst/sushi/"+str(page)
  print("第"+str(page)+"頁")
  response=urlopen(url)
  html=BeautifulSoup(response)

  for r in html.find_all("div",class_="list-rst__rst-data"):
    ja=r.find("a",class_="list-rst__rst-name-target cpy-rst-name")
    print(ja.text)

  page=page+1
```
